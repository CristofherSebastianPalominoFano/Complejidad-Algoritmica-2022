### Backtracking - N Reinas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tablero = [4, 7, 2, 6, 1, 0, 3, 5]

for row, col in enumerate(tablero):
  print(f"Hay una reina en la fila {row} columna {col}")
```

```python
def draw(tablero):
  n = len(tablero)
  b = np.zeros((n, n, 3), dtype=int)
  b += [255, 128, 80]

  b[::2, ::2] = [255, 225, 120]
  b[1::2, 1::2] = [255, 225, 120]

  _, ax = plt.subplots()
  ax.imshow(b)
  for row, col in enumerate(tablero):
    ax.text(col, row, u"\u265b", fontsize=200/n, va="center", ha="center")
  ax.set(xticks=[], yticks=[])
```

```python
draw(tablero)
```

```python
def validar(tablero, row, col):
  for row_i in range(row):
    col_i = tablero[row_i]
    delta = row - row_i
    if col in [col_i, col_i + delta, col_i - delta]:
      return False

  return True
```

```python
# El uso del assert en Python nos permite verificar que una determinada condición sea True, y de no serlo, 
# se lanzará una excepción.

tablero = [1, 3, -1, -1]
assert validar(tablero, 2, 0) == True
draw(tablero)
```

```python
#Posiciones validas en la fil,col = 0,1 y 1,3
#Posicion no valida de una 3ra reina en el tablero, en la fil,col = 2,1
tablero = [1, 3, -1, -1]
assert validar(tablero, 2, 1) == False
draw(tablero)
```

```python
#Posiciones validas en la fil,col = 0,1 y 1,3
#Posicion no valida de una 3ra reina en el tablero, en la fil,col = 2,2
tablero = [1, 3, -1, -1]
assert validar(tablero, 2, 2) == False
draw(tablero)
```

```python
#Generalizamos una funcion para las N Reinas

def nqueens(tablero, row):
  n = len(tablero)
  if row == n:
    draw(tablero)
  else:
    for col in range(n):
      if validar(tablero, row, col):
        tablero[row] = col
        nqueens(tablero, row+1)
```
